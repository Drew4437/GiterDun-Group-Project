# Read main dataset in dataframe

```python
# read Main Starter File with Funding and weather data combined (CSV)
starter_data = pd.read_csv("../startup_datasets/final_data1.csv")
starter_data.head()
```

```python
# Create a scatter plot for the Acquirer data set (x=# of acquisitions, y=$ total acquisition cost, color for categories)
starter_data.columns

# ******** NEED TO CHANGE ACTUAL CSV FILE to eliminate "-" in Acquired_Price column

### Steps to create graphic
# 1) Create dataframe with required data (Acquirer, Acquisition Count, Acquisition Total, etc)
# 2) Create metrics / values for the scatter plot, and create scatter plot
```

```python
# 1) Create dataframe with required data (Acquirer, Acquisition Count, Acquisition Total, etc)
# First pull only the data needed from starter_data df: Acquirer, Acquired_Price
acquirers = pd.DataFrame(starter_data.loc[starter_data['status'] == 'acquired', ["Acquirer", "Acquired_Price"]])

# Find the values requested
acquirers_grouped = acquirers.groupby("Acquirer")
acquirers_grouped_count = acquirers_grouped["Acquired_Price"].count() 
acquirers_grouped_avpx = acquirers_grouped["Acquired_Price"].mean()
acquirers_grouped_total = acquirers_grouped["Acquired_Price"].sum()

# Create table to present the values requested
acquirers_grouped_table = pd.DataFrame({"Acquisition Count": acquirers_grouped_count,
                                        "Average Acquisition Price": acquirers_grouped_avpx,
                                        "Total Acquisition Spending": acquirers_grouped_total})
acquirers_table = acquirers_grouped_table.sort_values(by='Acquisition Count', ascending=False)
acquirers_table.head()
```

```python
# 2) Create metrics / values for the scatter plot, and create scatter plot
x_acquirer = acquirers_table['Acquisition Count']
y_acquirer = acquirers_table['Average Acquisition Price']/1000000
s_acquirer = acquirers_table['Total Acquisition Spending']/250000000


fig1 = plt.figure()
ax1 = fig1.add_subplot()

ax1.scatter(x_acquirer, y_acquirer, s=s_acquirer, linewidths=1, edgecolor='black')

textstr = 'Note:\nCircle size correlates with total acquisition spending ($250 million)\n'
# print(textstr)
plt.text(27,5000,textstr, fontsize=12)
plt.title("Acquisitions Data")
plt.xlabel("Total Number of Acquisitions (per Acquirer)")
plt.ylabel("Average Acquisition Price ($ millions)")
plt.grid(True)

# # Save Figure
plt.savefig("../Images/Acquisitions_Full_Set.png", edgecolor='none', bbox_inches = 'tight', orientation='landscape')

# # Using plotly (interactive graphic)
# fig1 = go.Figure(data=go.Scatter(
#     x=x_acquirer,
#     y=y_acquirer,
#     mode='markers',
#     marker=dict(size=s_acquirer)
# ))


fig1.show()
```

```python
fig2 = go.Figure(data=go.Scatter(
    x = x_acquirer,
    y = y_acquirer,
    mode='markers',
    marker=dict(
        size=s_acquirer,
        color=x_acquirer, #set color equal to a variable
        colorscale='Viridis', # one of plotly colorscales
        showscale=True
    )
))
fig2.update_layout(
    title='Acquisition Data',
    xaxis=dict(
        title='Acquisitions per Acquirer',
        gridcolor='white',
        gridwidth=2,
    ),
    yaxis=dict(
        title='Average Acquisition Price ($ millions)',
        gridcolor='white',
        gridwidth=2,
    ),
    paper_bgcolor='rgb(243, 243, 243)',
    plot_bgcolor='rgb(243, 243, 243)',
)
font=dict(
               family="'Oswald', sans-serif",
               size=12,
               color="#7f7f7f")
fig2.show()
```

```python
# Narrow data set to only focus on those companies that have acquired more than 7 (serial acquirers)
serial_acquirers = pd.DataFrame(acquirers_grouped_table.loc[acquirers_grouped_table["Acquisition Count"] >= 7, ["Acquisition Count", "Average Acquisition Price", "Total Acquisition Spending"]])
serial_acquirers['Total Acquisition Spending'].sum()
```

```python
# Serial Acquirers: Acquisition Count
fig3 = go.Figure()

fig3.add_trace(go.Barpolar(
    r=serial_acquirers["Acquisition Count"],
    theta=serial_acquirers.index,
    name='Acquisition Count',
    marker_color='rgb(158,154,200)'
))

fig3.update_traces(text=serial_acquirers.index)
fig3.update_layout(
    title='Serial Acquirers: Acquisition Count',
    font_size=12,
    legend_font_size=12,
    polar_angularaxis_rotation=90,
    
)

font=dict(
               family="'Oswald', sans-serif",
               size=12,
               color="#7f7f7f")
# plt.savefig("../Images/Acquisitions_Serial_Acquirers_Count.png", edgecolor='none', bbox_inches = 'tight', orientation='landscape')

fig3.show()
```

```python
# Serial Acquirers: Average Acquisition Price
fig4 = go.Figure()

fig4.add_trace(go.Barpolar(
    r=serial_acquirers["Average Acquisition Price"],
    theta=serial_acquirers.index,
    name='Average Price ($ millions)',
    marker_color='rgb(158,154,200)'
))

fig4.update_traces(text=serial_acquirers.index)
fig4.update_layout(
    title='Serial Acquirers: Average Acquisition Price ($)',
    font_size=12,
    legend_font_size=12,
    polar_angularaxis_rotation=90,
    
)
font=dict(
               family="'Oswald', sans-serif",
               size=12,
               color="#7f7f7f")
# plt.savefig("../Images/Acquisitions_Serial_Acquirers_AvgPx.png", edgecolor='none', bbox_inches = 'tight', orientation='landscape')

fig4.show()
```

```python
# Serial Acquirers: Total Acquisitions Spending

fig5 = go.Figure()

fig5.add_trace(go.Barpolar(
    r=serial_acquirers["Total Acquisition Spending"],
    theta=serial_acquirers.index,
    name='Total Spend ($ billions)',
    marker_color='rgb(158,154,200)'
))

fig5.update_traces(text=serial_acquirers.index)
fig5.update_layout(
    title='Serial Acquirers: Total Acquisition Spending ($ billions)',
    font_size=12,
    legend_font_size=12,
    polar_angularaxis_rotation=90,
    
)
font=dict(
               family="'Oswald', sans-serif",
               size=12,
               color="#7f7f7f")
# plt.savefig("../Images/Acquisitions_Serial_Acquirers_TotalSpend.png", edgecolor='none', bbox_inches = 'tight', orientation='landscape')

fig5.show()
```
